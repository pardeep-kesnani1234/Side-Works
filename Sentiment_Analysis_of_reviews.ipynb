```python
polar1 = []
for index, row in  packaging[['Review']].iterrows():
  sentiment_analyzed = sentiment_intensity.polarity_scores(row['Review'])
  polar1.append(sentiment_analyzed)
polarities1 = pd.Series(polar1)
packaging['polarity'] = polarities1.values
```

```python
polar2 = []
for index, row in  shipping[['Review']].iterrows():
  sentiment_analyzed = sentiment_intensity.polarity_scores(row['Review'])
  polar2.append(sentiment_analyzed)
polarities2 = pd.Series(polar2)
shipping['polarity'] = polarities2.values
```

```python
polar3 = []
for index, row in  quality[['Review']].iterrows():
  sentiment_analyzed = sentiment_intensity.polarity_scores(row['Review'])
  polar3.append(sentiment_analyzed)
polarities3 = pd.Series(polar3)
quality['polarity'] = polarities3.values
```

```python
for i in  quality["polarity"]:
  print(i)
```

```python
def sentiments_name(dict_, new_keys):
    """
     new_keys: type List(), must match length of dict_
    """

    # dict_ = {oldK: value}
    # d1={oldK:newK,} maps old keys to the new ones:  
    d1 = dict( zip( list(dict_.keys()), new_keys) )

          # d1{oldK} == new_key 
    return {d1[oldK]: value for oldK, value in dict_.items()}
```

```python
for index,val in  enumerate(packaging['polarity']):
  packaging['polarity'][index]=  sentiments_name(val,['NEGATIVE ','NEUTRAL', 'POSITIVE', 'COMPOUND'])
  print(packaging['polarity'][index])
```

```python
for index,val in  enumerate(shipping['polarity']):
  shipping['polarity'][index]=  sentiments_name(val,['NEGATIVE ','NEUTRAL', 'POSITIVE', 'COMPOUND'])
  print(shipping['polarity'][index])
```

```python
for index,val in  enumerate(quality['polarity']):
  quality['polarity'][index]=  sentiments_name(val,['NEGATIVE ','NEUTRAL', 'POSITIVE', 'COMPOUND'])
  print(quality['polarity'][index])
```

## Generate word cloud

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline

text1 = packaging.Review.values

text2 = shipping.Review.values

text3 = quality.Review.values



# adding movie script specific stopwords
stopwords = set(STOPWORDS)
stopwords.add("https")
stopwords.add("xa0")
stopwords.add("xa0'")
stopwords.add("bitly")
stopwords.add("bit")
stopwords.add("ly")
stopwords.add("twitter")
stopwords.add("pic")

wordcloud1 = WordCloud(
    background_color = 'black',
    width = 1000,
    height = 500,
    stopwords = stopwords).generate(str(text1))

wordcloud2 = WordCloud(
    background_color = 'yellow',
    width = 1000,
    height = 500,
    stopwords = stopwords).generate(str(text2))


wordcloud3 = WordCloud(
    background_color = 'orange',
    width = 1000,
    height = 500,
    stopwords = stopwords).generate(str(text3))    
```

```python
plt.imshow(wordcloud1, interpolation='bilinear')
plt.axis("off")
plt.rcParams['figure.figsize'] = [20, 20]
```

```python
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
plt.rcParams['figure.figsize'] = [15, 15]
```
